## **Simple Exponential Smoothing**

```python
# Import the necessary modules
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
from sklearn.metrics import mean_squared_error

# Create a simple exponential smoothing model
model = SimpleExpSmoothing(train['price'])

# Fit the model to the training data
model_fit = model.fit(smoothing_level=0.2, optimized=False)

# Make predictions on the test data
predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1)

# Plot the predictions
plt.figure(figsize=(12, 6))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(predictions, label='Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Simple Exponential Smoothing')
plt.legend()
plt.show()

# Calculate the mean squared error
simple_rmse = np.sqrt(mean_squared_error(test['price'], predictions))
print('Root Mean Squared Error:', simple_rmse)
```

```python
model_fit.summary()
```

## **Double (or) Holt's Exponential Smoothing Method**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Create a double exponential smoothing model
model = ExponentialSmoothing(train['price'], trend='additive', seasonal='additive')

# Fit the model to the training data
model2_fit = model.fit()

# Make predictions on the test data
predictions = model_fit.predict(start=len(train), end=len(train) + len(test) - 1)

# Plot the predictions
plt.figure(figsize=(12, 6))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(predictions, label='Predictions')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Double Exponential Smoothing')
plt.legend()
plt.show()

# Calculate the mean squared error
double_rmse = np.sqrt(mean_squared_error(test['price'], predictions))
print('Root Mean Squared Error:', double_rmse)
```

```python
model2_fit.summary()
```

## **Holts winter (or) Triple Exponential Smoothing Method**

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
model1 = ExponentialSmoothing(train['price'], trend='add', seasonal='add', seasonal_periods=12)
model1_fit = model1.fit()
yhat = model1_fit.forecast(len(test))
plt.figure(figsize=(15,7))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(yhat, label='Exponential Smoothing')
plt.legend(loc='best')
plt.show()
# prompt: check for rmse value

triple_rmse=np.sqrt(mean_squared_error(test['price'],yhat))
print("The root mean squared error is: ",triple_rmse)
```

```python
model1_fit.summary()
```

## **ARIMA MODEL**

```python
from pmdarima import auto_arima

stepwise_fit = auto_arima(gp['price'],
                          start_p=1, start_q=1, max_p=3, max_q=3,
                          m=12, start_P=0, seasonal=True, d=None, D=1, trace=True,
                          error_action='ignore', suppress_warnings=True, stepwise=True)
```

```python
from statsmodels.tsa.arima.model import ARIMA, ARIMAResults
# Fit the ARIMA model with the selected order and seasonal order


model = ARIMA(train['price'], order=(1,0,0),seasonal_order=(1,1,0,12))
amodel_fit = model.fit()
amodel_fit.summary()

# Make predictions
start = len(train)
end = len(train)+len(test)-1
print(start,end)
pred_arima = amodel_fit.predict(start=start, end=end)


# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(pred_arima, label='ARIMA Model')
plt.legend(loc='best')
plt.show()

# Calculate the RMSE of the ARIMA model
arima_rmse = np.sqrt(mean_squared_error(test['price'], pred_arima))
print('ARIMA RMSE:', arima_rmse)
```

```python
amodel_fit.summary()
```

## **SARIMA MODEL**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python

# Fit the SARIMA model with the selected order and seasonal order


sarima_model = SARIMAX(train['price'], order=(1,0,0),seasonal_order=(1,1,0,12))
sarima_model = sarima_model.fit()
sarima_model.summary()

# Make predictions
start = len(train)
end = len(train)+len(test)-1
print(start,end)
pred_arima = sarima_model.predict(start=start, end=end)


# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(pred_arima, label='SARIMA Model')
plt.legend(loc='best')
plt.show()

# Calculate the RMSE of the ARIMA model
sarima_rmse = np.sqrt(mean_squared_error(test['price'], pred_arima))
print('SARIMA RMSE:', arima_rmse)
```

```python
sarima_model.summary()
```

### **NAIVE MODEL**

```python
# Create a naive model by taking the last value of the training set as the prediction for each time step in the test set.
naive_predictions = np.array([train['price'][-1]] * len(test))

# Plot the actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(train['price'], label='Train')
plt.plot(test['price'], label='Test')
plt.plot(naive_predictions, label='Naive Model')
plt.legend(loc='best')
plt.show()

naive_rmse = np.sqrt(mean_squared_error(test['price'], naive_predictions))
print('Naive RMSE:', naive_rmse)
```

## **Linear Regression Model**




```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Prepare the data
X = gp.index.to_numpy().reshape(-1, 1)
y = gp['price'].to_numpy()

# Train the model
lrmodel = LinearRegression()
lrmodel.fit(X, y)
X = X.astype(np.float64)

# Make predictions
lr_predictions = lrmodel.predict(X)

# Evaluate the model
from sklearn.metrics import mean_squared_error
lin_rmse = np.sqrt(mean_squared_error(y, lr_predictions))
print('lin_rmse:', lin_rmse)
```

```python
# line plot for lr_predictions

plt.figure(figsize=(10,5))
plt.plot(lr_predictions)
plt.xlabel("Time")
plt.ylabel("Gold Price")
plt.title("Linear Regression Model Predictions")
plt.show()
```
